# Lecture one, assuming t1 = t1, t2 = t2, x(t1) = xi, x(t2) = xf

$$J(x) = \int_{t_1}^{t_2}\, L[x(t),x'(t),t]  $$

Assume the following:<br>
$ x(t) = x^{*}(t) + \epsilon y(t)$ <br>
$ x^{'}(t) = x^{'*}(t) + \epsilon y^{'}(t)$

Substituting:

$$J(x) = \int_{t_1}^{t_2}\, L[x^{*}(t) + \epsilon y(t),x^{'*}(t) + \epsilon y^{'}(t),t]  $$

The neccessary condition for optimization (We evaluate at epsilon = 0 since it is where the optimal solution exist):

$$
\left. \frac{\partial J}{\partial \epsilon} \right|_{\epsilon = 0} = 0
$$

$$\frac{\partial J}{\partial \epsilon} = \int_{t_1}^{t_2}\, \frac{\partial}{\partial \epsilon} L[x^{*}(t) + \epsilon y(t),x^{'*}(t) + \epsilon y^{'}(t),t]  $$

$$
\frac{\partial L}{\partial \epsilon} = \frac{\partial L}{\partial x}\frac{\partial x}{\partial \epsilon} + \frac{\partial L}{\partial x'}\frac{\partial x'}{\partial \epsilon}
$$

$$
\frac{\partial L}{\partial \epsilon} = \frac{\partial L}{\partial x}y(t) + \frac{\partial L}{\partial x'}y'(t)
$$

In this case y(t) and y'(t) are both dependent, so for example, y(t) = 3 yeilds y'(t) = 0 and we can make no conclusion on the $\frac{\partial L}{\partial x'}$ term, because of this we do integration by parts to get multiple y(t) terms instead of a y(t) and y'(t)

$$\int_{t_1}^{t_2} \frac{\partial L}{\partial x'}\dot{y}\, dt = \left[\frac{\partial L}{\partial x'}y\right]_{t_1}^{t_2} - \int_{t_1}^{t_2} \frac{d}{dt}\frac{\partial L}{\partial x'} y\, dt$$

$$\left[\frac{\partial L}{\partial x'}y\right]_{t_1}^{t_2} = 0 \quad \text{since } y(t_1) = y(t_2) = 0$$

$$\int_{t_1}^{t_2} \left[\frac{\partial L}{\partial x} - \frac{d}{dt}\frac{\partial L}{\partial x'}\right]y\, dt = 0$$

$$\frac{\partial L}{\partial x} - \frac{d}{dt}\frac{\partial L}{\partial x'} = 0$$

## The Hamiltonian and Transversality Conditions

The Hamiltonian is defined as:

$$H = x'\frac{\partial L}{\partial x'} - L$$

For a conservative system (e.g. a spring), this is always equal to the total mechanical energy:

$$H = \frac{1}{2}m\dot{x}^2 + \frac{1}{2}kx^2 = E$$

This is **conserved** along the optimal path when $L$ has no explicit time dependence.

---

## Transversality Conditions

Starting from:

$$\delta J = \left[L - x'\frac{\partial L}{\partial x'}\right]_{t_f}\delta t_f + \frac{\partial L}{\partial x'}\bigg|_{t_f}\delta x_f = 0$$

| Fixed | Free | Transversality Condition |
|-------|------|--------------------------|
| $x_f$ | $t_f$ | $H\big\|_{t_f} = 0$ |
| $t_f$ | $x_f$ | $\frac{\partial L}{\partial x'}\bigg\|_{t_f} = 0$ (momentum = 0) |

---

## Physical Interpretation

- $H = 0$ at $t_f$ means the system has **zero total energy** at the stopping time
- For a pure conservative spring this forces $t_f \to \infty$ since $H = E \neq 0$ always
- The condition is only physically meaningful when energy can **dissipate** (damping, control inputs etc.) so that $H$ can actually reach zero at a finite $t_f$
- This is why transversality conditions are most natural in **optimal control** rather than classical mechanics

$$-\lambda = 3 \implies \lambda = -3$$

Therefore:
$$x_1 = \frac{3}{2}, \quad x_2 = -\frac{3}{2}$$

Optimal point:
$$\underline{x}^* = \begin{pmatrix} \frac{3}{2} \\ -\frac{3}{2} \end{pmatrix}, \quad f(\underline{x}^*) = \left(\frac{3}{2}\right)^2 + \left(-\frac{3}{2}\right)^2 = \frac{9}{4} + \frac{9}{4} = \frac{9}{2}$$

# Alternatively without using substitution:

$$J_{a}(x) = \frac{1}{2} \int_{0}^{t_f}\, [x^{2}(t) + u^{2}(t)] dt + \int_{0}^{t_f}\, [\lambda (f(u,x)-x'(t))] dt $$
where $$f = -x + u$$

$$\lambda (t) = \lambda_0(t)$$

$$ L =  \frac{1}{2}(x^{2}(t) + u^{2}(t))$$

Then:

$$J_{a}(x) = \int_{0}^{t_f}\, [\lambda (f(u,x)-x'(t))] dt $$

where

$$H =  L + \lambda f(u,x)$$

Then: 

$$J_{a}(x) = \int_{0}^{t_f}\, [H - \lambda x'(t)] dt $$

First order optimality condition: 
$$\delta J_{a}(x) = \delta \int_{0}^{t_f}\, [H - \lambda x'(t)] dt = 0 $$

$$\delta J_{a}(x) = \int_{0}^{t_f}\, [\delta H - \delta \lambda x'(t) - \lambda \delta x'(t)] dt = 0 $$

$$\delta J_{a}(x) = \int_{0}^{t_f}\, [\frac{\partial H}{\partial u}\delta u + \frac{\partial H}{\partial {x}}\delta {x} + \frac{\partial H}{\partial \lambda}\delta \lambda  - \delta \lambda x'(t) - \lambda \delta x'(t)] dt = 0 $$

$$\delta H = \frac{\partial H}{\partial u}\delta u + \left(\frac{\partial H}{\partial \underline{x}} + \dot{\underline{\lambda}}\right)\delta \underline{x} + \left(\frac{\partial H}{\partial \underline{\lambda}} - \dot{\underline{x}}\right)\delta \underline{\lambda} = 0$$

Since $\delta u$, $\delta \underline{x}$, $\delta \underline{\lambda}$ are independent, each term must vanish separately:

$$\frac{\partial H}{\partial u} = 0$$

$$\frac{\partial H}{\partial \underline{x}} + \dot{\underline{\lambda}} = 0 \implies \dot{\underline{\lambda}} = -\frac{\partial H}{\partial \underline{x}}$$

$$\frac{\partial H}{\partial \underline{\lambda}} - \dot{\underline{x}} = 0 \implies \dot{\underline{x}} = \frac{\partial H}{\partial \underline{\lambda}}$$

Given:
$$H = \frac{1}{2}(x^2 + u^2) + \lambda(-x + u)$$

Stationarity condition:
$$\frac{\partial H}{\partial u} = u + \lambda = 0 \implies u = -\lambda$$

Costate equation:
$$\dot{\lambda} = -\frac{\partial H}{\partial x} = -\frac{\partial}{\partial x}\left[\frac{1}{2}x^2 + \lambda(-x+u)\right] = -(x - \lambda) = -x + \lambda$$

$$\implies \dot{\lambda} = -x + \lambda$$

State equation:
$$\dot{x} = \frac{\partial H}{\partial \lambda} = -x + u$$

Substituting $u = -\lambda$:
$$\dot{x} = -x - \lambda$$

**Writing as a matrix system:**
$$\begin{pmatrix} \dot{x} \\ \dot{\lambda} \end{pmatrix} = \begin{pmatrix} -1 & -1 \\ -1 & 1 \end{pmatrix} \begin{pmatrix} x \\ \lambda \end{pmatrix}$$
