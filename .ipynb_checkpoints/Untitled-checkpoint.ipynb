```python
# ============================================================
# Brachistochrone Optimal Control — Shooting Method
# ============================================================
import numpy as np
from scipy.integrate import solve_ivp
from scipy.optimize import fsolve

g = 10.0

def dynamics(t, z):
    x, y, v, lamx, lamy, lamv = z

    A = lamx * v
    B = lamy * v + lamv * g
    theta = np.arctan2(A, B)               # Eq. 2.7

    dxdt   = v * np.sin(theta)             # Eq. 2.1
    dydt   = v * np.cos(theta)             # Eq. 2.2
    dvdt   = g * np.cos(theta)             # Eq. 2.3
    dlamxdt = 0.0                          # Eq. 2.4
    dlamydt = 0.0                          # Eq. 2.5
    dlamvdt = -lamx*np.sin(theta) - lamy*np.cos(theta)  # Eq. 2.6

    return [dxdt, dydt, dvdt, dlamxdt, dlamydt, dlamvdt]


def integrator(lamx0, lamy0, lamv0, tf):
    z0  = [0, 0, 0, lamx0, lamy0, lamv0]
    sol = solve_ivp(dynamics, [0, tf], z0, rtol=1e-10, atol=1e-10)
    return sol.y, sol.t


# ------------------------------------------------------------------
# WHY lamv(0) IS PINNED TO -1/g
# ------------------------------------------------------------------
# The full Pontryagin Hamiltonian is H_total = 1 + λ·f = 0 everywhere.
# At t=0, v=0, so A=0 and B = λv(0)·g.
# The physical control starts at θ=0 (bead drops straight down).
# Therefore:
#   H_total(0) = 1 + λv(0)·g·cos(0) = 1 + λv(0)·g = 0
#   => λv(0) = -1/g = -0.1
#
# arctan2 cannot see this because it computes arctan2(0, negative) = π
# (the wrong branch). Pinning λv(0) = -1/g removes that ambiguity
# and reduces the system from 4 unknowns to 3.
# ------------------------------------------------------------------

lamv0_fixed = -1.0 / g   # pinned by H(0+) = -1 condition

def error(guess):
    lamx0, lamy0, s = guess
    tf = np.exp(s)
    z, _ = integrator(lamx0, lamy0, lamv0_fixed, tf)

    r1 = z[0, -1] - 2   # x(tf) = 2
    r2 = z[1, -1] - 2   # y(tf) = 2
    r3 = z[5, -1]        # λv(tf) = 0  (transversality, v(tf) free)

    return np.array([r1, r2, r3])


# Initial guess — good starting point near the analytic cycloid answer
guess = np.array([0.15, 0.056, np.log(0.8165)])

solution, info, ier, msg = fsolve(error, guess, xtol=1e-10, full_output=True)

lamx0_sol = solution[0]
lamy0_sol = solution[1]
tf        = np.exp(solution[2])

z, t = integrator(lamx0_sol, lamy0_sol, lamv0_fixed, tf)

print("=" * 50)
print("  SHOOTING METHOD SOLUTION")
print("=" * 50)
print(f"lambda_x(0) = {lamx0_sol:.10f}")
print(f"lambda_y(0) = {lamy0_sol:.10f}")
print(f"lambda_v(0) = {lamv0_fixed:.10f}  [pinned = -1/g]")
print(f"tf          = {tf:.10f}")
print(f"x(tf)       = {z[0,-1]:.10f}")
print(f"y(tf)       = {z[1,-1]:.10f}")
print(f"lambda_v(tf)= {z[5,-1]:.2e}")
print(f"Converged:    {ier==1}  — {msg}")

# ------------------------------------------------------------------
# Verify H = -1 everywhere (with branch-corrected theta)
# ------------------------------------------------------------------
lx, ly, lv, v_arr = z[3], z[4], z[5], z[2]
A = lx * v_arr
B = ly * v_arr + lv * g
theta_plot = np.arctan2(A, B)

# Branch correction: ensure cos(θ) >= 0 (physical: v must increase from rest)
neg = np.cos(theta_plot) < 0
theta_plot[neg] -= np.sign(np.where(theta_plot[neg] == 0, 1.0, theta_plot[neg])) * np.pi

H = lx*v_arr*np.sin(theta_plot) + ly*v_arr*np.cos(theta_plot) + lv*g*np.cos(theta_plot)

print()
print("=" * 50)
print("  VERIFICATION")
print("=" * 50)
print(f"Mean H along path   = {np.mean(H):.10f}  (must be -1.0)")
print(f"Max |H+1| deviation = {np.max(np.abs(H+1)):.2e}  (must be ~0)")
print(f"theta(0)            = {np.degrees(theta_plot[0]):.4f}°  (expect 0°)")
print(f"theta(tf)           = {np.degrees(theta_plot[-1]):.4f}°  (expect ~69.1°)")

# Analytic comparison (cycloid)
R_cyc    = 1.145834
phi_f    = 2.412011
tf_exact = np.sqrt(R_cyc / g) * phi_f
print()
print(f"Analytic (cycloid) tf = {tf_exact:.10f}")
print(f"Numeric            tf = {tf:.10f}")
print(f"Error                 = {abs(tf - tf_exact):.2e}")
```
